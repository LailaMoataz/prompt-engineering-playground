# Prompt Engineering Experiments
**Objective:** Explore prompt effectiveness for different LLM tasks.

## Experiment 1: Summarization
**Prompt:** "Summarize the following text in 3 bullet points."

**Input Text:**
"Artificial intelligence is transforming many industries, from healthcare to finance..."

**Simulated Output:**
- AI is impacting multiple sectors.
- Healthcare and finance see major changes.
- AI adoption is increasing globally.

**Notes:** This prompt produces concise, clear summaries. Can try different word limits or styles.


## Experiment 2: Rewrite for ClarityExperiment 2: Rewrite for Clarity

**Prompt:** "Rewrite the following paragraph so it is clear and concise."

**Input Text:** "The process of AI integration into business operations is complicated and requires careful consideration."

**Simulated Output:** "Integrating AI into business is complex and needs careful planning."

**Notes:** Prompt works well; could add tone specification (formal/informal).


## Experiment 3: Generate Study Questions

**Prompt:** "Create 5 multiple-choice questions from the following text."

**Input Text:** "The water cycle consists of evaporation, condensation, precipitation, and collection."

**Simulated Output:**
1. What is the first stage of the water cycle?  
   a) Condensation  b) Evaporation  c) Precipitation  d) Collection  

   **Answer:** b) Evaporation

2. ... (and so on for 5 questions)

**Notes:** Clear, easy-to-read questions. Could experiment with difficulty levels.
